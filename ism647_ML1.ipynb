## Inspect the Data: Nulls
<p>
Look for features with a high number of null values and consider how to address.
<p>
Removed rows with missing values. They represented only 12% of the total 

```python
# print number of missing values per feature. filter to only those with missing values

missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]

# add column for percent of total missing values for each feature
missing_values = missing_values.to_frame()
missing_values.columns = ['missing_count']
missing_values['percent_missing'] = missing_values['missing_count'] / df.shape[0]

# print missing values per feature to inspect
print(missing_values)
```

```python
# create a dataframe with rows from features that have missing values

missing_df = df[df.isnull().any(axis=1)]

# print head of missing_df

missing_df.head(50)
```

## Inspect the Data: Variance
<p>
Look for features representing with a low variance and consider how to address.
<p>
4 features were type float and had variance of 0. From data dictionary, these features were actually categorical features stored as float. Changed to int.

```python
# find variance for each numerical feature in dataframe
numerical_features = df.select_dtypes(include=[np.number])

# calculate variance for each numerical feature as an integer
variance = numerical_features.var().astype(int)

# select features with a variance of 0
variance = variance[variance == 0]

# add column for the data type of each feature
variance = variance.to_frame()
variance.columns = ['variance']
variance['data_type'] = df[variance.index].dtypes

# select features with a variance of 0 and type of float64
variance = variance[(variance['variance'] == 0) & (variance['data_type'] == 'float64')]

# print variance for each numerical feature
print(variance)
```

```python
# in df, select features in variance dataframe and change their data type to int64
df[variance.index] = df[variance.index].astype('int64')

# print data type of the features in variance dataframe to confirm they have been changed
print(df[variance.index].dtypes)
```
